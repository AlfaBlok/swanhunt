# Risk management using Kelly Cristerion and information theory

Shannon defined Information as the log of the reverse probability of an event. The more unlikely an event, the more information it contains. The more likely an event, the less information is necesary to describe it, and therefore communicate it. 

<!-- Here we add the formula of information, as I(x) = log (1/p_x)  in latex -->

$I(x) = \log \left(\frac{1}{p_x}\right)$


```python
import matplotlib.pyplot as plt
import numpy as np

# Define the range of probabilities
p = np.linspace(0.01, 1, 100)

# Calculate the Shannon information
I = -np.log2(p)

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(p, I, label='I(x) = -log2(p)')
plt.xlabel('Probability')
plt.ylabel('Shannon Information')
plt.title('Shannon Information vs Probability')
# we hide the grid lines
plt.grid(False)

plt.legend()
# plt.grid(True)
plt.show()
```

Entropy is defined as the expected value of information

$H(X) = -\sum_{i=1}^{n} p(x_i) \log(p(x_i))$

```python
# now we plot the entropy as a function of probability in the binary case
H = -p*np.log2(p) - (1-p)*np.log2(1-p)
plt.figure(figsize=(10, 6))
plt.plot(p, H, label='H(x) = -p*log2(p) - (1-p)*log2(1-p)')
plt.xlabel('Probability')
plt.ylabel('Entropy')
plt.title('Entropy vs Probability')
plt.grid(False)
plt.legend()
plt.show()
```

The entropy or H is the minimum possible amount of information required to describe without errors stochastic process X.

Entropy works in such a way that the more stable a process is, less information is required to describe its state, and therefore has a lower entropy value. The inverse is also true. If a process is very unstable a lot of information will be required on average to describe it's state.

Entropy tells us how much information is required to communicate a process.


In the presence of noise, the maximum transmission capacity of information over a channel of a certain maximumn Bandwith is:

$C = B \log_2(1 + S/N)$

This means that we cannot possible transfer more information that 

$f* = \frac{E[R] - r_f}{\sigma^2}$

$f* = \frac{0.03 - 0}{0.10^2} = 3$

When in communications we find the Optimal Channel capacity, we are told the ceiling of the information we can transmit of the channel without loss.

If we transfer more information the message is corrupted.

Same in Kelly criterion, if we try to invest (transmit) more than the channel can support, we degrade growth (communication).
